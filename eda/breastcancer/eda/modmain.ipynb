```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns 
from sklearn import preprocessing
from sklearn.model_selection import train_test_split 
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report
import pickle
```

```python
dataf = pd.read_csv("G:\\siddhant\\mlds\\breastcancer\\eda\\data.csv")
dataf.head()
```

```python
dataf.isnull().sum()
```

```python
str_cols = dataf.columns[dataf.dtypes==object] 
dataf[str_cols] = dataf[str_cols].fillna('.')
dataf.describe()
```

```python
dataf.columns 
```

```python
listf = [] 
for c in dataf.columns: 
    print(c,"->" ,dataf) 
    listf.append(c) 
listf
```

```python
def labelEncoding(data): 
    dataset = pd.DataFrame(dataf) 
    feature_dict ={} 
      
    for feature in dataset: 
        if dataset[feature].dtype==object: 
            le = preprocessing.LabelEncoder() 
            fs = dataset[feature].unique() 
            le.fit(fs) 
            dataset[feature] = le.transform(dataset[feature]) 
            feature_dict[feature] = le 
              
    return dataset 
  
labelEncoding(dataf)
```

```python
dataf = dataf.drop(['Unnamed: 32'], axis =1)
dataf
```

```python
dataf['diagnosis'] = dataf['diagnosis'].map({ 'M': 1, 'B': 0 })
```

```python
dataf
```

```python
X = dataf.drop(['diagnosis'], axis=1)
y= dataf['diagnosis']
```

```python
scaler = StandardScaler()
X = scaler.fit_transform(X)
```

```python
X_train,x_test,y_train, y_test = train_test_split(X,y,test_size=0.33, random_state=42)
```

```python
# train the model
model = LogisticRegression()
model.fit(X_train, y_train)
```

```python
# test model
y_pred = model.predict(x_test)
print('Accuracy of our model: ', accuracy_score(y_test, y_pred))
print("Classification report: \n", classification_report(y_test, y_pred))
```
